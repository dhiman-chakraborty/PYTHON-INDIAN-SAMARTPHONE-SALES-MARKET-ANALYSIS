```python
import numpy as np
import pandas as pd
```

```python
customers = pd.read_csv(r'D:\PYTHON PROJECTS\INDIAN SAMARTPHONE SALES MARKET\CLEANED DATASETS\cleaned_Customer_data.csv',encoding='latin1')
Order_item = pd.read_csv(r'D:\PYTHON PROJECTS\INDIAN SAMARTPHONE SALES MARKET\CLEANED DATASETS\cleaned_Order_items_data.csv',encoding='latin1')
Orders = pd.read_csv(r'D:\PYTHON PROJECTS\INDIAN SAMARTPHONE SALES MARKET\CLEANED DATASETS\cleaned_Orders_data.csv',encoding='latin1')
Payments = pd.read_csv(r'D:\PYTHON PROJECTS\INDIAN SAMARTPHONE SALES MARKET\CLEANED DATASETS\cleaned_Payments_data.csv',encoding='latin1')
Products = pd.read_csv(r'D:\PYTHON PROJECTS\INDIAN SAMARTPHONE SALES MARKET\CLEANED DATASETS\cleaned_Products_data.csv',encoding='latin1')
Review = pd.read_csv(r'D:\PYTHON PROJECTS\INDIAN SAMARTPHONE SALES MARKET\CLEANED DATASETS\cleaned_Review_data.csv',encoding='latin1')

print('Customer table:\n',customers.head())
print('Order_item table:\n',Order_item.head())
print('Orders table:\n',Orders.head())
print('Payments table:\n',Payments.head())
print('Products table:\n',Products.head())
print('Review table:\n',Review.head())
```

```python
# checking null values
print(customers.isnull().sum())
print(Order_item.isnull().sum())
print(Orders.isnull().sum())
print(Payments.isnull().sum())
print(Products.isnull().sum())
print(Review.isnull().sum())
```

```python
# merger with order column
df = pd.merge(left=Orders,right=customers,on='customer_id',how='left')
df = pd.merge(left=df,right=Order_item,on='order_id',how='left')
df = pd.merge(left=df,right=Products,left_on='product_id_x',right_on='product_id',how='left')
df = pd.merge(left=df,right=Payments,on='order_id',how='left')
df = pd.merge(left=df,right=Review,on=['customer_id','product_id'],how='left')

print('main order table\n',df)
print(list(df.columns))
```

```python
df = df.drop(columns=['product_id_y','city_x','payment_method_y','quantity_y','price_y'])

df = df.rename(columns={'customer_id_x':'customer_id','product_id_x':'product_id','quantity_x':'quantity',
                        'price_x':'price','payment_method_x':'payment_method','city_y':'city'})

print('main order table\n',df)
print(list(df.columns))

print(df.info())
print(df.isnull().sum())
```
