```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

                                        ###### Importing & Cleaning Customer data ######

```python
Order_item = pd.read_csv(r'D:\PYTHON PROJECTS\INDIAN SAMARTPHONE SALES MARKET\DATASETS\Customers.csv',encoding='latin1')
print(Order_item.head(10))
print(Order_item.info())
```

```python
# checking null value
print(Order_item.isnull().sum())
```

```python
# clean customer dataset
Order_item = Order_item.drop_duplicates(subset=['customer_id'])    # makeing customer_id column unique

Order_item['age'] = Order_item['age'].astype(str).str.strip()
Order_item['age'] = Order_item['age'].replace({'Thirty':30})   # replace age column string 'Thirty' & 'Nan' To 30 & 0 
Order_item['age'] = Order_item['age'].fillna(0)                    # fill missing value with 0

Order_item['age'] = pd.to_numeric(Order_item['age'],errors='coerce')

Order_item['city'] = Order_item['city'].str.title().str.strip()    # make city name in proper form

Order_item['signup_date'] = pd.to_datetime(Order_item['signup_date'],format='mixed',dayfirst=True,errors='coerce') # covert & fixed mixed datetime
Order_item['signup_date'] = Order_item['signup_date'].dt.strftime('%d-%m-%y')     # covert in indian date format

Order_item['gender'] = Order_item['gender'].str.title().str.strip()      
Order_item['gender'] = Order_item['gender'].fillna('Not Disclosed')     # fixed gender column in proper form and replace missing values with 'Not disclosed'

Order_item['income'] = pd.to_numeric(Order_item['income']).fillna(0)
Order_item['loyalty_score'] = Order_item['loyalty_score'].replace({'High':4}).fillna(0)   # replace 'High' to 4 & missing values with 0 in loyalty_score
Order_item['loyalty_score'] = pd.to_numeric(Order_item['loyalty_score'])     # convert loyalty_score to numeric

Order_item['device_type'] = Order_item['device_type'].str.title().str.strip().fillna('Not Specific')  # fixed device_type to proper form & replace missing values with 'Not Specific'

print(Order_item.head(20))
print(Order_item.info())
```

                                            ##### Importing & Cleaning Order_item data ######

```python
Order_item = pd.read_csv(r'D:\PYTHON PROJECTS\INDIAN SAMARTPHONE SALES MARKET\DATASETS\Order_items.csv',encoding='latin1')
print(Order_item.head(10)
print(Order_item.info())
```

```python
Order_item = Order_item.drop_duplicates(subset=['order_item_id'])   # make sure 'order_item_id' contains unique value

Order_item['quantity'] = Order_item['quantity'].replace({'one':1}).fillna(1)  # covert 'one' to 1 & missing value to 1
Order_item['quantity'] = pd.to_numeric(Order_item['quantity'])      # covert quantity column to numeric

Order_item['unit_price'] = Order_item['unit_price'].astype(str).str.replace(r'[^\d]', '', regex=True)    # clean unite price digits
Order_item['unit_price'] = pd.to_numeric(Order_item['unit_price'], errors='coerce')      # convert to numeric 

Order_item['discount'] = Order_item['discount'].astype(str).str.replace(r'[^\d]','',regex=True)   # fixed 'Ã¢âÂ¹100' values
Order_item['discount'] = Order_item['discount'].replace({'5%':5,'10%':10,'100':10}).infer_objects(copy=False).fillna(0)     # convert '%' values to digit
Order_item['discount'] = pd.to_numeric(Order_item['discount'],errors='coerce')                    # conert to numeric

Order_item['tax'] = Order_item['tax'].replace({'5%':5,'12%':12,'18%':18}).fillna(0)           # convert '%' values to digit
Order_item['tax'] = pd.to_numeric(Order_item['tax'],errors='coerce')                          # conert to numeric

Order_item['shipping_fee'] = (Order_item['shipping_fee']
                              .astype(str).str.strip()
                              .replace({'Free':0})
                              .str.replace(r'[^\d]','',regex=True).fillna(0))           # covert free to 0 & 'Ã¢âÂ¹100' to 100
Order_item['shipping_fee'] = pd.to_numeric(Order_item['shipping_fee'],errors='coerce')     # covert to numeric

Order_item['item_status'] = Order_item['item_status'].str.title().str.strip().fillna('Not Yet Confirmed')    # fixed in proper form and 
                                                                                                             # missing value to 'Not Yet Confirmed'

Order_item['Total Bill'] = ((Order_item['unit_price']*Order_item['quantity'])
                            +((Order_item['unit_price']*Order_item['quantity'])*Order_item['tax']/100)
                            -(((Order_item['unit_price']*Order_item['quantity'])
                            +((Order_item['unit_price']*Order_item['quantity'])*Order_item['tax']/100))*Order_item['discount']/100)
                            +Order_item['shipping_fee'])

print(Order_item.head(20))
print(Order_item.info())
```

                                             #### Importing & Cleaning Products data #####

```python
products = pd.read_csv(r'D:\PYTHON PROJECTS\INDIAN SAMARTPHONE SALES MARKET\DATASETS\Products.csv',encoding='latin1')

print(products.head(50))
print(products.info())
```

```python
print(products.isnull().sum())

print(products['product_id'].is_unique)         # checking 'product_id' contains unique values
```

```python
products['brand'] = products['brand'].str.strip().str.upper()

products['model'] = products['model'].str.strip()

products['price'] = products['price'].astype(str).str.strip().replace(r'[^\d]','',regex=True)      # remove unecessary thing & clean
products['price'] = pd.to_numeric(products['price'],errors='coerce')               # convert it to numeric

products['rating'] = products['rating'].str.strip().replace({'Five':5}).fillna(0)       # replace 'Five' with 5
products['rating'] = pd.to_numeric(products['rating'],errors='coerce')                  # convert to numeric

products['ram'] = products['ram'].str.strip().fillna('Not Specified') 

products['battery'] = products['battery'].str.strip().fillna('Not Specified') 

products['stock'] = products['stock'].str.strip().replace({'thirty':30}).fillna(0)
products['stock'] = pd.to_numeric(products['stock'],errors='coerce')

products['seller_city'] = products['seller_city'].str.strip().str.title()

print(products.head(50))
print(products.info())
print(products.isnull().sum())
```

```python
review = pd.read_csv(r'D:\PYTHON PROJECTS\INDIAN SAMARTPHONE SALES MARKET\DATASETS\Reviews.csv',encoding='latin1')

print(review.head(50))
print(review.info())
```

```python
print(review.isnull().sum())

print(review['review_id'].is_unique)     # checking 'review_id' is primary key or not
```

```python
review['rating'] = review['rating'].astype(str).str.strip().replace({'Five':5}).fillna(0)     # convert 'Five' to 5 & fill blanks with 0
review['rating'] = pd.to_numeric(review['rating'],errors='coerce').fillna(0)                                            # convert column datatype to numeric(float)

review['review_text'] = review['review_text'].str.strip().str.title().fillna('No Review')      # Fixed string column & fill blank with 'No Review'

review['review_date'] = pd.to_datetime(review['review_date'],format='mixed',dayfirst=True,errors='coerce')  # fixed date column

review['likes'] = review['likes'].astype(str).str.strip().replace({'twenty':20}).fillna(0)           # convert 'twenty' to 20 and fill blanks with 0
review['likes'] = pd.to_numeric(review['likes'],errors='coerce').fillna(0)

review['verified'] = review['verified'].astype(str).str.strip().fillna('Not Confirmed')        # fill blanks with 'Not Confirmed

review['device'] = review['device'].str.strip().str.title().fillna('No Device')              # fill blanks with no device

print(review.head(10))
print(review.info())
print(review.isnull().sum())
```
